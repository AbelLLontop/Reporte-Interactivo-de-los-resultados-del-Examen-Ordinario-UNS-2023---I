```python
def test(row):
    text_especialidad = row['Especialidad']
    next_position = row.name+1
    salir = 0
    while(salir!=-1):  
        if next_position<=(len(df)-1):
            next_row = df.iloc[next_position]
            if pd.isna(next_row['#']):
                text_especialidad+=(" "+next_row['Especialidad'].strip())
                next_position+=1
                
            else:
                salir = -1
        else:
            salir = -1
    row['Especialidad'] = text_especialidad
    return row
    
f1 = df.apply(test,axis='columns')
```

```python
f1['Especialidad'].unique().tolist()
```

```python
notNanIngreso = f1['Ingreso'].notna()
f2 = f1.loc[notNanIngreso]
```

```python
f2['Especialidad'].unique().tolist()
```

```python
def add_new_column(row):
    text_new_column = 'DESCARTADO'
    if row['Especialidad'].startswith('1'):
        text_separate = row['Especialidad'].split('2°')
        text_new_column = text_separate[1].strip()
        row['Especialidad'] = text_separate[0][2:].strip()
        
    row['Segunda Opcion'] = text_new_column
    return row
f3 = f2.apply(add_new_column,axis='columns')
f3[f3['Ingreso']=='Seg Opc']
```

```python
f4 = f3.set_index('#')
f4
```

```python
f5 = f4
f5.index = f5.index.astype(int)
f5
```

```python
f5.columns = ['CARNE',
              'APELLIDOS Y NOMBRES',
              'APTITUD ACADEMICA - BUENAS',
              'APTITUD ACADEMICA - MALAS',
              'APTITUD ACADEMICA PUNTAJE',
              'CONOCIMIENTOS - BUENAS',
              'CONOCIMIENTOS - MALAS',
              'CONOCIMIENTOS PUNTAJE',
              'PUNTAJE FINAL',
             'ESPECIALIDAD',
             'INGRESO',
             'SEGUNDA OPCION']
f5.head()
```
